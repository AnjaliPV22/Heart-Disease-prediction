```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('heart_.csv')
dataset.head()
```

```python
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler 
sc=StandardScaler()
x_train=sc.fit_transform(x_train)  #xscaled=x-mean/std: to make the data optimised or standardised
x_test=sc.transform(x_test)
```

```python
x_train
```

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
df=pd.DataFrame(y_pred,y_test)
df
```

```python
classifier.predict(sc.transform([[63,	1,	3,	145,	233,	1,	0,	150,	0,	2.3,	0,	0,	1]]))
```

```python
classifier.predict(sc.transform([[57,	0,	0,	120,	354,	0,	1,	163,	1,	0.6,	2,	0,	2	]]))
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score
cm=confusion_matrix(y_test,y_pred)
print(cm)

accuracy=accuracy_score(y_test,y_pred)
print("Accuracy",accuracy)

precision=precision_score(y_test,y_pred)
print("Precision",precision)

recall=recall_score(y_test,y_pred)
print("Recall",recall)
```
