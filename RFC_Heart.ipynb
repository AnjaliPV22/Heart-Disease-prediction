```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('heart_.csv')
dataset.head()
```

```python
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler 
sc=StandardScaler()
x_train=sc.fit_transform(x_train)  #xscaled=x-mean/std: to make the data optimised or standardised
x_test=sc.transform(x_test)
```

```python
x_train
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier(n_estimators=100,criterion='gini',max_depth=5)
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
df=pd.DataFrame(y_pred,y_test)
df
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```

```python
#select one tree from forest
from sklearn.tree import plot_tree
tree=classifier.estimators_[0]
#plot decision tree
plt.figure(figsize=(15,10))
plot_tree(tree,feature_names=['age',	'sex',	'cp',	'trestbps',	'chol',	'fbs',	'restecg',	'thalach',	'exang',	'oldpeak',	'slope',	'ca',	'thal'],class_names=['No','Yes'])
plt.title("Decision tree from Random Forest heart")
plt.show()
```
